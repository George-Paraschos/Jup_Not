```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

points1 = np.array([(0, 0), (70, 60), (140, 100), (350, 110), (420,130), (490,140), (350,80), (700,130), (1050,140), (1400,150), (1750,160), (2450,170), (3150,170), (350,100), (1050,160), (1750,200), (2800,180), (4550,190), (6300,170)])
x1 = points1[:,0]
y1 = points1[:,1]

x1_error=20
y1_error=700
x_error=20
y_error=700

points = np.array([(0, 0), (175, 20), (350, 50), (525, 100), (700,140), (875,160), (1225,190), (700,100), (1400,120), (1750,150), (5950,200), (12250,190), (1750,100), (3500,200), (5250,220), (7000,200), (8750,190), (12250,170), (15750,170), (22750,160), (26250,150)])
x = points[:,0]
y = points[:,1]

#label='CO(2-->1)'
#label='HI'



plt.plot(x1,y1,'bs',label='HI')
plt.plot(x,y,'g^',label='CO(2-->1)')
plt.errorbar(x,y, x_error, y_error, fmt ='none')
plt.errorbar(x1,y1, x1_error, y1_error, fmt ='none')

plt.legend(loc='lower right')


plt.xlabel('Distance (pc)')
plt.ylabel('Velocity (km/s)')
plt.title('Superposition of HI and CO(2-->1)')
plt.ylim(-0,250)
plt.xlim(-0.1,30000)
plt.savefig("CC.png")
```
