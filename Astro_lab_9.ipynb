```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

b=3000
G=4.3/1000
M=2.2*100000000000


x = np.arange(0.0, 5000.0, 0.1)
y = np.sqrt(x*x/(2*b*b)*(1-x*x/(4*b*b))*(1-x*x/(2*b*b))*G*M)


plt.plot(x,y)

plt.xlabel('Distance (pc)')
plt.ylabel('Velocity (km/s)')
plt.title('Rotational curve for the stars near the centre')
plt.ylim(-0,20000)
plt.xlim(-0,5000)
plt.savefig("RCC.png")
```
